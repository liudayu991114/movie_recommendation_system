Part I Create the Login Frame

```python
login = Frame(base, width = 450, height = 300)
login.place(x = 640, y = 360, anchor = 'center')

login_canvas = tk.Canvas(login, height = 150, width = 500)
pic_login = tk.PhotoImage(file = 'bg/welcome.png')
bg_login = login_canvas.create_image(0, 0, anchor = 'nw', image = pic_login)
login_canvas.place(x = 225, y = 0, anchor = 'n')

tk.Label(login, text = 'User name: ', font=('Arial', 12)).place(x = 0, y = 150, anchor = 'nw')
tk.Label(login, text = 'Password: ', font=('Arial', 12)).place(x = 0, y = 190, anchor = 'nw')

var_login_name = tk.StringVar()
var_login_name.set('first.last@example.com')
entry_login_name = tk.Entry(login, textvariable = var_login_name, font=('Arial', 12), width = 30)
entry_login_name.place(x = 120, y = 150, anchor = 'nw')

var_login_pwd = tk.StringVar()
entry_login_pwd = tk.Entry(login, textvariable = var_login_pwd, show = '*', font=('Arial', 12), width = 30)
entry_login_pwd.place(x = 120, y = 190, anchor = 'nw')
```

```python
#Define the command of Login Button
def login_f():

    global login_name
    login_name = var_login_name.get()
    login_pwd = var_login_pwd.get()
    
    try:
        with open('dataset/accounts_info.pickle', 'rb') as accounts_file:
            accounts_info = pickle.load(accounts_file)
    except FileNotFoundError:
        with open('dataset/accounts_info.pickle', 'wb') as accounts_file:
            accounts_info = {'admin':'admin'}
            pickle.dump(accounts_info, accounts_file)
    
    if login_name in accounts_info:
        if login_pwd == accounts_info[login_name]:
            try:
                user_path = 'dataset/users/{}.csv'
                user_file = user_path.format(login_name)
                with open(user_file, 'r') as user_data:
                    reader = csv.reader(user_data)
            except FileNotFoundError:
                path = 'dataset/users/{}.csv'
                user_file = path.format(login_name)
                with open(user_file, 'w', newline = '') as user_data:
                    writer = csv.writer(user_data)
                    row0 = ['title', 'rating']
                    writer.writerow(row0)
            finally:
                tk.messagebox.showinfo(title = 'Login Successful', message = 'How are you? ' + login_name)
                login.destroy()
                main_f()
        else:
            tk.messagebox.showerror(message = 'Wrong password, try again.')
    else:
        is_signup = tk.messagebox.askyesno(title = 'Welcome', message = 'Account does not exist. Sign up?')
        if is_signup:
            signup_f()
```

```python
#Define the command of Sign Up Button
#Create the Sign Up Window
def signup_f():
    
    try:
        with open('dataset/accounts_info.pickle', 'rb') as accounts_file:
            accounts_info = pickle.load(accounts_file)
    except FileNotFoundError:
        with open('dataset/accounts_info.pickle', 'wb') as accounts_file:
            accounts_info = {'admin':'admin'}
            pickle.dump(accounts_info, accounts_file)

    def sign_to_system():
        nn = new_name.get()
        np = new_pwd.get()
        npf = new_pwd_confirm.get()
        with open('dataset/accounts_info.pickle', 'rb') as accounts_file:
            accounts_info = pickle.load(accounts_file)
        if np != npf:
            tk.messagebox.showerror(title = 'Error', message = 'Password and Confirmed password must be the same')
        elif nn in accounts_info:
            tk.messagebox.showerror(title = 'Error', message = 'The user name already exists!')
        else:
            accounts_info[nn] = np
            with open('dataset/accounts_info.pickle', 'wb') as accounts_file:
                pickle.dump(accounts_info, accounts_file)
            tk.messagebox.showinfo(title = 'Welcome', message = 'You have successfuly Signed Up, Welcome to the World of Movies')
            signup.destroy()
    
    signup = tk.Toplevel(base)
    signup.geometry('350x200')
    signup.title('Sign Up')

    new_name = tk.StringVar()
    new_name.set('first.last@example.com')
    tk.Label(signup, text = 'User name: ', font=('Arial', 12)).place(x = 10, y = 20)
    entry_new_name = tk.Entry(signup, textvariable = new_name, font=('Arial', 12), width = 20)
    entry_new_name.place(x = 150, y = 20)

    new_pwd = tk.StringVar()
    tk.Label(signup, text = 'Password: ', font=('Arial', 12)).place(x = 10, y = 60)
    entry_usr_pwd = tk.Entry(signup, textvariable = new_pwd, show = '*', font=('Arial', 12), width = 20)
    entry_usr_pwd.place(x = 150, y = 60)

    new_pwd_confirm = tk.StringVar()
    tk.Label(signup, text = 'Confirm password: ', font=('Arial', 12)).place(x = 10, y = 100)
    entry_usr_pwd_confirm = tk.Entry(signup, textvariable = new_pwd_confirm, show = '*', font=('Arial', 12), width = 20)
    entry_usr_pwd_confirm.place(x = 150, y = 100)

    btn_confirm_sign_up = tk.Button(signup, text = 'Sign up', font=('Arial', 12), width = 8, height = 1, command = sign_to_system)
    btn_confirm_sign_up.place(x = 140, y = 150)
```

```python
#Create the buttons
btn_login = tk.Button(login, text = 'Login', font=('Arial', 12), width = 8, height = 1, command = login_f)
btn_login.place(x = 120, y = 230)
btn_sign_up = tk.Button(login, text = 'Sign up', font=('Arial', 12), width = 8, height = 1, command = signup_f)
btn_sign_up.place(x = 270, y = 230)
tk.Label(login, text = 'Input \'admin\' for both User Name and Password to test.', font=('Arial', 12)).place(x = 225, y = 280, anchor = 'n')
```

Part III.I Create the Giving Scores Toplevel

```python
#Create the Scoring Frame

def score_f():
    score = tk.Toplevel(base)
    score.title('Give Your Scores')
    score.geometry('1280x720')

    def confirm_score_f():
        movie_name = var_movie_name.get()
        movie_year = var_movie_year.get()
        movie_info = movie_name + '(' + movie_year + ')'
        user_path = 'dataset/users/{}.csv'
        user_file = user_path.format(login_name)
        with open(user_file, 'r') as user_data:
            reader = csv.reader(user_data)
            movies_in = dict(reader)
        if movie_info in movies_in:
            tk.messagebox.showinfo(title = 'Error', message = 'You have already scored this movie')
        else:
            with open(user_file, 'a+', newline = '') as user_data:
                    writer = csv.writer(user_data)
                    rown = [movie_info, movie_score]
                    writer.writerow(rown)
            tk.messagebox.showinfo(title = 'Rating Successful', message = 'Your score has been stored')

    def back_score_f():
        score.destroy()

    tk.Label(score, text = 'Movie Name: ', font=('Arial', 15)).place(x = 400, y = 200, anchor = 'nw')
    tk.Label(score, text = 'You must input the Primary Title of the movie in IMDB (the official English name of the movie).', font=('Arial', 12)).place(x = 400, y = 250, anchor = 'nw')
    var_movie_name = tk.StringVar()
    var_movie_name.set('Name of the movie')
    var_movie_year = tk.StringVar()
    var_movie_year.set('Year')
    entry_movie_name = tk.Entry(score, textvariable = var_movie_name, font=('Arial', 15), width = 24)
    entry_movie_name.place(x = 550, y = 200, anchor = 'nw')
    entry_movie_year = tk.Entry(score, textvariable = var_movie_year, font=('Arial', 15), width = 10)
    entry_movie_year.place(x = 820, y = 200, anchor = 'nw')

    tk.Label(score, text = 'Your Scores: ', font=('Arial', 15)).place(x = 400, y = 300, anchor = 'nw')
    l_score = tk.Label(score, bg = 'white', width = 34, text = 'You Score will be Shown Here', font=('Arial', 15))
    l_score.place(x = 550, y = 300, anchor = 'nw')

    def print_selection(v):
        l_score.config(text = 'Your Score for this Movie is ' + v)
        global movie_score
        movie_score = v

    s = tk.Scale(score, label = 'Choose Your Score', font=('Arial', 15), from_ = 0, to = 10, orient = tk.HORIZONTAL,
        length = 600, showvalue = 0, tickinterval = 1, resolution = 0.1,
        command = print_selection)
    s.place(x = 350, y = 380, anchor = 'nw')

    btn_score_confirm = tk.Button(score, text = 'Confirm', font=('Arial', 18), width = 12, height = 1, command = confirm_score_f)
    btn_score_confirm.place(x = 640, y = 520, anchor = 'center')

    btn_score_back = tk.Button(score, text = 'Back', font=('Arial', 15), width = 12, height = 1, command = back_score_f)
    btn_score_back.place(x = 20, y = 700, anchor = 'sw')

    btn_score_quit = tk.Button(score, text = 'Quit', font=('Arial', 15), width = 12, height = 1, command = quit_f)
    btn_score_quit.place(x = 1260, y = 700, anchor = 'se')
```

```python
def mlist_f():
    mlist = tk.Toplevel(base)
    mlist.title('View my Movie List')
    mlist.geometry('1280x720')

    tk.Label(mlist, text = 'The movie you select and your score are', font=('Arial', 15), width = 60).place(x = 640, y = 100, anchor = 'center')
    var_show = tk.StringVar()
    tk.Label(mlist, bg = 'white', font=('Arial', 15), width = 60, textvariable = var_show).place(x = 640, y = 150, anchor = 'center')

    def print_selection():
        print_value = m_list.get(m_list.curselection())
        var_show.set(print_value)

    btn_print = tk.Button(mlist, text = 'Print Selection', font=('Arial', 15), width = 12,
        height = 1, command = print_selection)
    btn_print.place(x = 640, y = 570, anchor = 'center')

    user_path = 'dataset/users/{}.csv'
    user_file = user_path.format(login_name)
    with open(user_file, 'r', encoding = 'utf-8') as user_data:
            reader = csv.reader(user_data)
            movies_in = list(reader)
    movies_shown = tk.StringVar()
    movies_shown.set(movies_in)
    m_list = tk.Listbox(mlist, listvariable = movies_shown, font=('Arial', 12), width = 50, height = 16)
    m_list.place(x = 640, y = 200, anchor = 'n')

    def back_mlist_f():
        mlist.destroy()

    btn_score_back = tk.Button(mlist, text = 'Back', font=('Arial', 15), width = 12, height = 1, command = back_mlist_f)
    btn_score_back.place(x = 20, y = 700, anchor = 'sw')

    btn_score_quit = tk.Button(mlist, text = 'Quit', font=('Arial', 15), width = 12, height = 1, command = quit_f)
    btn_score_quit.place(x = 1260, y = 700, anchor = 'se')
```

Part III.III Create the Recommendation Toplevel

```python
#Create the Recommendation Frame
i = 0
j = 0
def recom_f():
    recom = tk.Toplevel(base)
    recom.title('Recommendations')
    recom.geometry('1280x720')

    global i
    i = 0
    global j
    j = 0
    
    def back_recom_f():
        global i
        global j
        i = 0
        j = 0
        recom.destroy()
    
    def gene_recom_f():
        user_path = 'dataset/users/{}.csv'
        user_file = user_path.format(login_name)
        global movies_result
        movies_result = rd_rec(user_file, 20, 2, 0 , 19)
        global i
        recom_result.set(str(movies_result[i][0]))
        movie_id = get1imdbid(movies_result[i][0])
        getpng(movie_id)
        pic_path = 'pic folder/{}.png'
        pic_file = pic_path.format(movie_id)
        global show_pic
        show_pic = tk.PhotoImage(file = pic_file)
        print_image = image_canvas.create_image(0, 0, anchor = 'nw', image = show_pic)
        i = i + 1

    def next_recom_f():
        global i
        if i <= len(movies_result) - 1:
            recom_result.set(str(movies_result[i][0]))
            movie_id = get1imdbid(movies_result[i][0])
            getpng(movie_id)
            pic_path = 'pic folder/{}.png'
            pic_file = pic_path.format(movie_id)
            global show_pic
            show_pic = tk.PhotoImage(file = pic_file)
            print_image = image_canvas.create_image(0, 0, anchor = 'nw', image = show_pic)
            i = i + 1
        else:
            i = 0

    tk.Label(recom, text = 'I Need Personalized Recommendations', font=('Arial', 18)).place(x = 320, y = 70, anchor = 'center')

    recom_result = tk.StringVar()
    show_result = tk.Label(recom, bg = 'white', text = '', textvariable = recom_result, font=('Arial', 15), width = 40)
    show_result.place(x = 320, y = 120, anchor = 'center')

    image_canvas = tk.Canvas(recom, bg = 'white', height = 281, width = 190)
    global show_pic
    show_pic = tk.PhotoImage(file = 'bg/default.png')
    print_image = image_canvas.create_image(0, 0, anchor = 'nw', image = show_pic)
    image_canvas.place(x = 320, y = 170, anchor = 'n')
    
    btn_g_recom = tk.Button(recom, text = 'Generate Personalized Sequence', font=('Arial', 15), width = 30, height = 1, command = gene_recom_f)
    btn_g_recom.place(x = 320, y = 520, anchor = 'center')
    btn_n_recom = tk.Button(recom, text = 'Next One', font=('Arial', 15), width = 15, height = 1, command = next_recom_f)
    btn_n_recom.place(x = 320, y = 600, anchor = 'center')

    #following are codes for radomized recommendations

    def gene_recom_r_f():
        user_id = ''
        film_type = var_film_type.get()
        film_actor = var_film_actor.get()
        film_age = (var_film_year_start.get(), var_film_year_end.get())
        num = var_film_num.get()
        global movies_result_r
        recommender = Recommendation()
        movies_result_r = recommender.recommend_with_conditions(user_id, film_type, film_actor, film_age, num=10)
        global j
        recom_r_result.set(str(movies_result_r[j]))
        j = j + 1
    
    def next_recom_r_f():
        global j
        if j <= var_film_num.get() - 1:
            recom_r_result.set(str(movies_result_r[j]))
            j = j + 1
        else:
            j = 0

    tk.Label(recom, text = 'I Need Randomized Recommendations', font=('Arial', 18)).place(x = 960, y = 70, anchor = 'center')

    recom_r_result = tk.StringVar()
    show_r_result = tk.Label(recom, bg = 'white', text = '', textvariable = recom_r_result, font=('Arial', 15), width = 40)
    show_r_result.place(x = 960, y = 120, anchor = 'center')

    tk.Label(recom, text = 'Film Type', font=('Arial', 15)).place(x = 800, y = 180, anchor = 'nw')
    var_film_type = tk.StringVar()
    entry_film_type = tk.Entry(recom, textvariable = var_film_type, font=('Arial', 15), width = 20)
    entry_film_type.place(x = 910, y = 180, anchor = 'nw')

    tk.Label(recom, text = 'Actor', font=('Arial', 15)).place(x = 800, y = 230, anchor = 'nw')
    var_film_actor = tk.StringVar()
    entry_film_actor = tk.Entry(recom, textvariable = var_film_actor, font=('Arial', 15), width = 20)
    entry_film_actor.place(x = 910, y = 230, anchor = 'nw')

    tk.Label(recom, text = 'Period', font=('Arial', 15)).place(x = 800, y = 280, anchor = 'nw')
    var_film_year_start = tk.StringVar()
    entry_film_year_start = tk.Entry(recom, textvariable = var_film_year_start, font=('Arial', 15), width = 8)
    entry_film_year_start.place(x = 910, y = 280, anchor = 'nw')

    tk.Label(recom, text = 'to', font=('Arial', 15)).place(x = 1010, y = 280, anchor = 'nw')
    var_film_year_end = tk.StringVar()
    entry_film_year_end = tk.Entry(recom, textvariable = var_film_year_end, font=('Arial', 15), width = 8)
    entry_film_year_end.place(x = 1040, y = 280, anchor = 'nw')

    tk.Label(recom, text = 'Number', font=('Arial', 15)).place(x = 800, y = 330, anchor = 'nw')
    var_film_num = tk.StringVar()
    var_film_num.set('10')
    entry_film_num = tk.Entry(recom, textvariable = var_film_num, font=('Arial', 15), width = 20)
    entry_film_num.place(x = 910, y = 330, anchor = 'nw')

    tk.Label(recom, text = 'Currently this part has some Bugs.\nBut Personalized Recommendation is Perfect!', font=('Arial', 15)).place(x = 960, y = 400, anchor = 'n')

    tk.Label(recom, text = 'It will take some time to generate the recomendation sequence.', font=('Arial', 15)).place(x = 640, y = 650, anchor = 'n')

    btn_g_recom_r = tk.Button(recom, text = 'Generate Randomized Sequence', font=('Arial', 15), width = 30, height = 1, command = gene_recom_r_f)
    btn_g_recom_r.place(x = 960, y = 520, anchor = 'center')
    btn_n_recom_r = tk.Button(recom, text = 'Next One', font=('Arial', 15), width = 15, height = 1, command = next_recom_r_f)
    btn_n_recom_r.place(x = 960, y = 620, anchor = 'center')

    btn_recom_back = tk.Button(recom, text = 'Back', font=('Arial', 15), width = 12, height = 1, command = back_recom_f)
    btn_recom_back.place(x = 20, y = 700, anchor = 'sw')

    btn_recom_quit = tk.Button(recom, text = 'Quit', font=('Arial', 15), width = 12, height = 1, command = quit_f)
    btn_recom_quit.place(x = 1260, y = 700, anchor = 'se')
```
